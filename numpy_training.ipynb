19. Создайте случайный двумерный массив, элементы которого распределены нормально с ненулевым средним и неединичным стандартным отклонением.
Вычтите из каждого элемента среднее и поделите этот элемент на его стандартное отклонение.
Вычислите среднее и стандартное отклонение получившегося массива.

```python
import numpy as np
import random

# ar = np.asarray([random.randint(0, 5) for i in range(100)])
ar = np.random.normal(5, 2, 1000)
# ar = np.reshape(ar, (10, 10))
print(ar)
print(np.std(ar))
print(np.mean(ar))
```

```python
ar = (ar - np.mean(ar)) / np.std(ar)
print(ar)
print(np.std(ar))
print(np.mean(ar))
```
